```python
import numpy as np

import pandas as pd

from sklearn import linear_model

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
df
```

```python
df.rename(columns={'per capita income (US$)': 'per_capita_income'}, inplace=True)
```

```python
plt.scatter(df.year,df.per_capita_income, marker='+')
```

```python
new_df = df.drop('per_capita_income',axis='columns')

new_df
```

```python
per_capita_income = df.per_capita_income
```

```python
model = linear_model.LinearRegression()
model.fit(new_df, per_capita_income)
```

```python
model.predict([[2007]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
%calculating value by linear regression equation

2007*828.46507522-1632210.7578554575
```

Predicting per capita income for the year of 2020.

```python
model.predict([[2020]])
```

Taking all the prediction to a csv file

```python
all_year = pd.read_csv('years.csv')
```

```python
all_year
```

```python
all_predicted_values = model.predict(all_year)
```

```python
all_year['per capita income'] = all_predicted_values
```

```python
all_year.to_csv('predictions.csv')
```
